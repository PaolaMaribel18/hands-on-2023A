(a) Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```python
import numpy as np

# Generate an array of 10 random integers between 1 and 100
arr = np.random.randint(1, 101, size=10)

# Calculate the mean
mean = np.mean(arr)

# Calculate the standard deviation
std = np.std(arr)

print("Array:", arr)
print("Mean:", mean)
print("Standard Deviation:", std)
```

```python
import random
import math

# Generate an array of 10 random integers between 1 and 100
array = []
for _ in range(10):
    array.append(random.randint(1, 100))

# Calculate the mean
mean = sum(array) / len(array)

# Calculate the standard deviation
variance = sum((x - mean) ** 2 for x in array) / len(array)
std = math.sqrt(variance)

#Print the results
print("\033[1m" + "Results"+ "\033[0m")
print("Array:", array)
print("Mean:", mean)
print("Standard Deviation:", std)
```

(b) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate
the sum of each row and column.

```python
import random
import numpy as np

# Create a 2-dimensional array of random integer values
array = np.random.randint(1, 101, size=(3, 4))

# Calculate the sum of each row
row_sums = np.sum(array, axis=1)

# Calculate the sum of each column
column_sums = np.sum(array, axis=0)

print("Array:")
print(array)
print("Row Sums:", row_sums)
print("Column Sums:", column_sums)
```

```python
import random

# Create a 2-dimensional array of random integer values
array = []
for _ in range(3):
    row = []
    for _ in range(4):
        row.append(random.randint(1, 100))
    array.append(row)

# Calculate the sum of each row
row_sums = []
for row in array:
    row_sum = sum(row)
    row_sums.append(row_sum)

# Calculate the sum of each column
column_sums = []
for j in range(4):
    column_sum = 0
    for i in range(3):
        column_sum += array[i][j]
    column_sums.append(column_sum)

print("Array:")
for row in array:
    print(row)

print("Row Sums:", row_sums)
print("Column Sums:", column_sums)
```

(c) Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the
DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than
or equal to 30.

```python
import pandas as pd

# Create a DataFrame
data = {
    "Name": ["Paola", "Alejandro", "Daniela", "Cristian", "Ivan"],
    "Age": [23, 8, 31, 38, 47],
    "City": ["Quito", "Ambato", "Cuenca", "Guayaquil", "Manta"]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n")
df
```

```python
# Filter the DataFrame based on age, in this case 30 y/o
filtered_df = df[df["Age"] >= 30]
print("Filtered DataFrame:")
filtered_df
```

(d) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the
columns in the DataFrame.

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/kaggle/input/servicearea/ServiceArea.csv')


# Calculate the mean of a column
mean_value = df['VersionNum'].mean()

# Calculate the median of a column
median_value = df['VersionNum'].median()

# Calculate the mode of a column
mode_value = df['VersionNum'].mode().values[0]
print(df.head())

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

(e) Create a scatter plot of random x and y values between 1 and 100.

```python
import matplotlib.pyplot as plt
import random

# Generate random x and y values
x = [random.randint(1, 100) for _ in range(100)]
y = [random.randint(1, 100) for _ in range(100)]

# Create the scatter plot
plt.scatter(x, y)

# Set the plot title and labels
plt.title("Scatter Plot of Random Values")
plt.xlabel("X")
plt.ylabel("Y")

# Display the plot
plt.show()
```
